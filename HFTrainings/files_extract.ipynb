Converting the Jsonl dataset into txt such that it displays functions line and line while also preserving the "\n \t" characters.

```python
from datasets import Dataset, load_dataset, DatasetDict

ds_train = load_dataset("code_search_net", "java", split="train")
ds_test = load_dataset("code_search_net", "java", split="test")
ds_valid = load_dataset("code_search_net", "java", split="validation")
raw_datasets = DatasetDict(
    {
        "train": ds_train.shuffle().select(range(4000)), #train_size)), # "train": ds_train,  # .shuffle().select(range(50000)),
        "test": ds_test.shuffle().select(range(500)),
        "valid": ds_valid.shuffle().select(range(500)) # "valid": ds_valid,  # .shuffle().select(range(500))
    }
)
```

```python
# Open a new text file in write mode
with open("whole_func_strings.txt", "w") as file:
    # Iterate through each sample in the test split
    for sample in raw_datasets["test"]:
        # Write the "whole_func_string" to the file, preserving newline and tab characters
        file.write(repr(sample["whole_func_string"]) + "\n")
```

```python
# Open a new text file in write mode
with open("whole_func_strings1.txt", "w") as file:
    # Iterate through each sample in the test split
    for sample in raw_datasets["test"]:
        # Get the whole_func_string
        func_string = sample["whole_func_string"]
        # Remove the starting and ending single quotes or double quotes
        func_string = func_string.strip("'\"")
        # Write the cleaned whole_func_string to the file, preserving newline and tab characters
        file.write(repr(sample["whole_func_string"]) + "\n")
```

Convert Java code extracted from java corpus to single line 

```python
# Define the input and output file paths
input_file_path = 'java.txt'
output_file_path = 'output_java.txt'

# Open the input file for reading
with open(input_file_path, 'r') as input_file:
    # Read lines from the input file
    lines = input_file.readlines()

# Modify each line as required
modified_lines = []
for line in lines:
    # Strip leading and trailing whitespace from the line
    line = line.strip()
    # Check if the line starts with 'public' or 'private'
    if line.startswith(('public', 'private')):
        # If yes, add a newline before the line
        modified_lines.append('\n' + line)
    else:
        # If not, concatenate with the previous line if it's not empty
        if modified_lines and modified_lines[-1]:
            modified_lines[-1] += ' ' + line
        else:
            modified_lines.append(line)

# Join the modified lines into a single string
modified_content = '\n'.join(modified_lines)

# Write the modified content to the output file
with open(output_file_path, 'w') as output_file:
    output_file.write(modified_content)

print("Modified content has been saved to:", output_file_path)
```

```python
# Define the input and output file paths
input_file_path = 'java.txt'
output_file_path = 'output_java.txt'

# Open the input file for reading
with open(input_file_path, 'r') as input_file:
    # Read lines from the input file
    lines = input_file.readlines()

# Modify each line as required
modified_lines = []
function_started = False
for line in lines:
    # Strip leading and trailing whitespace from the line
    line = line.strip()
    # Check if the line starts with 'public' or 'private'
    if line.startswith(('public', 'private')):
        if function_started:
            modified_lines.append("}\n")
        # If yes, add a newline before the line
        modified_lines.append(line + " {")
        function_started = True
    elif line.strip().endswith('}'):
        # If the line ends with '}', add newline after the line
        modified_lines.append(line)
        function_started = False
    else:
        # If not, concatenate with the previous line if it's not empty
        if modified_lines and modified_lines[-1]:
            modified_lines[-1] += '\\n\\t' + line
        else:
            modified_lines.append(line)

# Join the modified lines into a single string
modified_content = '\n'.join(modified_lines)

# Write the modified content to the output file
with open(output_file_path, 'w') as output_file:
    output_file.write(modified_content)

print("Modified content has been saved to:", output_file_path)
```
